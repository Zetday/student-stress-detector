```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

# **Data Gathering**

```python
df = pd.read_csv("../data/processed/university_student_stress_clean.csv")
df.head()
```

# **Penilaian Distribusi Data**

```python
df.describe()
```

```python
numeric_data = df.select_dtypes(include=["float64","int64"])
n_cols = 2
n_rows = -(-len(numeric_data.columns) // n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15,25))

axes = axes.flatten()

for i, feature in enumerate(numeric_data.columns):
    sns.histplot(data=df[feature], bins=15, ax=axes[i], kde=True)
    axes[i].set_title(f"Feature of {feature}")
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])


plt.tight_layout()
plt.show()
```
