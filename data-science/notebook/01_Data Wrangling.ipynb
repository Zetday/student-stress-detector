# **Data Cleaning**

### Memperbaiki Invallid Data type

```python
# Memperbaiki Tipe data Salah denga dikonversi menjadi numerik
df['Caffeine_Intake_mg'] = pd.to_numeric(df['Caffeine_Intake_mg'], errors='coerce')
```

```python
"""
Memperbaiki Invalid pada kolom Study_Hours,Class_Attendance,Assignment_Load,Caffeine_Intake_mg,Stress_Score
Karena Memiliki nilai negatif yang seharusnya positif
"""
column = ["Study_Hours","Class_Attendance","Assignment_Load","Caffeine_Intake_mg","Stress_Score"]
for feature in column:
  df[feature] = df[feature].abs()
```

```python
# Nilai Age = 31 berada di luar rentang valid dataset (18–24), sehingga dianggap sebagai kemungkinan typo dan dikoreksi menjadi 21.
df['Age'] = df['Age'].replace(31, 21)
# Satuan Persentase hanya memiliki Rentang (1-100)
df["Class_Attendance"] = df["Class_Attendance"].replace(105, df["Class_Attendance"].median())

# Nilainya diluar Rentang valid dataset (1-10)
df["Exam_Frequency"] = df["Exam_Frequency"].replace(15, df["Exam_Frequency"].median())
df["Assignment_Load"] = df["Assignment_Load"].replace(12, df["Assignment_Load"].median())
df["Peer_Pressure"] = df["Peer_Pressure"].replace(11, df["Peer_Pressure"].median())
```

### Memperbaiki Data Tidak Konsisten

```python
# Memperbaiki data tidak konsisten pada kolom kategorikal
categorical_columns = df.select_dtypes(include=["object", "string", "category"]).columns

for feature in categorical_columns:
    df[feature] = (
        df[feature]
        .astype("string")
        .str.strip()
        .str.title()
    )
```

### Membersihkan Missing Value

```python
# Membersihkan missing value pada data numerik
numeric_column = df.select_dtypes(include=["int64", "float64"]).columns
df[numeric_column] = df[numeric_column].fillna(df[numeric_column].median())
```

```python
# Membersihkan missing value pada Data Kategorial
for feature in categorical_columns:
  df[feature] = df[feature].fillna(df[feature].mode()[0])
```

```python
# Hapus Duplikat Data
df.drop_duplicates(inplace=True)
```
